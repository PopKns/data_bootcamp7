```python
from talib import RSI, EMA, ATR
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = xauuse_futures_data
print(data)
```

```python
close = data['close'].values
print(close)
```

```python
ma50 = EMA(close, timeperiod=50)
ma100 = EMA(close, timeperiod=100)
ma200 = EMA(close, timeperiod=200)
```

```python
plt.figure()
plt.plot(data['close'].values, label='Price')
plt.plot(ma50,label='EMA 50')
plt.plot(ma100,label='EMA 100')
plt.plot(ma200,label='EMA 200')
plt.legend(loc='best')
```

```python
rsi14 = RSI(close, timeperiod=14)
print(rsi14)
```

```python
atr14 = ATR(data['high'], data['low'], data['close'], timeperiod=14)
print(atr14)
```

```python
data['ma50'] = ma50
data['ma100'] = ma100
data['ma200'] = ma200
data['rsi14'] = rsi14
data['atr14'] = atr14
```
